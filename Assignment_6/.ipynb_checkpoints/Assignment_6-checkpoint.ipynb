```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/weather.csv')
df.head()
```

```python
sns.distplot(df['MinTemp'])
```

```python
sns.distplot(df['MinTemp'],kde=False,rug=True)
```

```python
sns.distplot(df['MinTemp'],hist=False,rug=True)
```

```python
sns.kdeplot(df['MinTemp'],shade=True)
```

```python
sns.kdeplot(df['MaxTemp'],label='original')
sns.kdeplot(df['MaxTemp'],bw=0.2,label='bw=0.2')
sns.kdeplot(df['MaxTemp'],bw=4,label='bw=2')
plt.legend()
```

```python
df.plot(kind='scatter',x='Temp9am',y='Temp3pm')
```

```python
sns.jointplot(x='Pressure9am',y='Pressure3pm',data=df)
```

```python
sns.jointplot(x='Pressure9am',y='Pressure3pm',data=df,kind='hex',color='k')
```

```python
sns.jointplot(x='Pressure9am',y='Pressure3pm',data=df,kind='kde')
```

```python
sns.pairplot(data=df,vars=['Temp9am','Temp3pm','Pressure9am','Pressure3pm'],hue='RainTomorrow')
```

```python
sns.stripplot(x = 'RainTomorrow', y='Rainfall', data=df)
```

```python
sns.swarmplot(x = 'RainTomorrow', y='Rainfall', data=df)
```

```python
sns.boxplot(x = 'RainTomorrow', y='Rainfall', data=df)
```

```python
sns.boxplot(x = 'RainTomorrow', y='Rainfall', data=df,hue='RainToday',dodge=True)
```

```python
sns.barplot(x = 'RainToday', y='Rainfall', data=df)
```

```python
sns.barplot(x = 'RainToday', y='Rainfall', data=df,hue='WindGustDir')
```

```python
sns.countplot(x='WindGustDir',data=df)
```
