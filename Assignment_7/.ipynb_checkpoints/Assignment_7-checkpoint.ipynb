<h3><b>Data Fields Explanation</b></h3><br>
Survived: "1" for Survived and "0" for Not Survived<br>
Pclass: Classes 1,2 and 3<br>
Name: Name of the person<br>
Sex: Gender of person<br>
Age: Age of peron<br>
Siblings/Spouses Aboard: Number of siblings or spouses aboard<br>
Parents/Children Aboard: Number of parents or children aboard<br>
Fare: The fare for travelling<br>

```python
print(df['Survived'][df['Survived']==1].count())
print(df['Pclass'][df['Pclass']==1].count())
print(df['Pclass'][df['Pclass']==2].count())
print(df['Pclass'][df['Pclass']==3].count())
print(df['Sex'][df['Sex']=='male'].count())
print(df['Sex'][df['Sex']=='female'].count())
print(df['Age'][df['Age']<25].count())
print(df['Age'][df['Age']>25][df['Age']<50].count())
print(df['Age'][df['Age']>50].count())
```

<b>5. Multivariate Analysis - Plot the multivariate graphs and find the relationship between the variables. Explain your findings and inference of the graphs.</b>

```python
num_features = list(set(df.columns)-set(['Name','Sex']))
cor_mat = df[num_features].corr()
sns.heatmap(cor_mat)
```

```python
sns.jointplot(x = 'Pclass' , y = 'Fare' , data=df , kind='scatter')
```

```python
sns.pairplot(df[num_features])
```

```python
sns.lmplot('Pclass', 'Fare', data=df,\
          hue='Survived', fit_reg=False)
```
