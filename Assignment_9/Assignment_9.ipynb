```python
import re
g = df.apply(lambda x:"".join(re.findall(', M[\w]+',x.Name)),axis=1)
df['Title'] = g
df.head()
```

# 5) Impute the missing values using .fillna() method.

```python
df = df.fillna(df.mean()).round(2)
df
```

# 6) Create two new columns “Categorical_age” and “Categorical_fare”, where you will put people that are within a certain range of age or fare in the same bin. 
You can do this by using the pandas function qcut() to bin your numerical data. (This is done because there is a range of ages and fares, so there might be fluctuations in those numbers that don't reflect patterns in the data, which might be noise.)
(https://pandas.pydata.org/pandas-docs/stable/generated/pandas.qcut.html) 

```python
binsAge = [0,25,50,75,100]
binsFare = [1,50,100,150,200,250,300,350,400]
X = ['0-25','25-50','50-75','75-100']
Y = ['1-50','50-100','100-150','150-200','200-250','250-300','300-350','350-400']
df['Categorical_age'] = pd.cut(df['Age'],binsAge,labels=X)
df['Categorical_fare'] = pd.cut(df['Fare'],binsFare,labels=Y)
df
```

# 7) Drop columns Cabin, Name, Passengerid, Ticket, Age, Fare, SibSp, Parch.

# 8) Transform all variables into numeric ones using get_dummies().

```python
df = pd.get_dummies(df, columns=['Survived'])
df
```

# 9) Now we need training set and testing set. Use train_test_split() to split your dataset. 
X(Features) will be the dataset values without the column “Survived” and y(Targets) will be your “Survived” column values.

```python
x= np.array(df[['Age', 'Fare', 'Parch', 'PassengerId', 'Pclass',
     'SibSp']])
y= np.array(df['Survived_0.0'])
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

# 10) Use logistic regression by passing your X_train and y_train.

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train,Y_train)
```

```python
y_pred = reg.predict(X_test)
```

# 11) Print the accuracy score. 

```python
acc = reg.score(X_test, Y_test)
print(acc)
```
